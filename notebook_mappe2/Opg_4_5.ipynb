# Sammenheng mellom Temperatur og Relativ Fuktighet

Denne visualiseringen utforsker forholdet mellom temperatur (°C) og relativ fuktighet (%). Vi bruker et scatterplot med regresjonslinje for å undersøke om det finnes en lineær sammenheng mellom de to variablene.

### Hvorfor denne typen plott?
- Punktene representerer individuelle målinger.
- Regresjonslinjen (rød) indikerer trenden i datasettet – altså hvordan fuktigheten typisk endres med temperaturen.
- Dette kan være nyttig for å forstå fysiske eller værrelaterte sammenhenger, som at varmere luft ofte holder på mer fuktighet.

### Hva vi kan tolke:
- En negativ skråning antyder at høyere temperaturer gir lavere relativ fuktighet.
- En positiv skråning betyr at fuktigheten øker med temperaturen.

### Tolkning av grafen:
Den røde linja går nedover fra venstre mot høyre, noe som betyr at det er en negativ sammenheng mellom temperatur og relativ fuktighet.
Dette gir fysisk mening fordi varm luft kan holde på mer vanndamp enn kald luft, så når temperaturen øker, kan den relative fuktigheten synke selv om mengden vann i lufta er konstant.





```python
# Oppretter et objekt med sti til datasettet
plotter = VærDataPlotter("../data/R_Sammenslaaing_uten_feil.csv")

# Kaller metoder for å lage grafer – velg det du vil se:
plotter.tegn_temp_vs_fuktighet()
```

# Korrelasjonsheatmap for vær- og miljødata

Denne visualiseringen viser sammenhengen (korrelasjon) mellom ulike numeriske værvariabler, som temperatur, lufttrykk, nedbør, relativ fuktighet og skydekke.

### Hva betyr korrelasjon?
- En verdi nær +1 betyr at to variabler øker sammen (positiv sammenheng).
- En verdi nær -1 betyr at når én variabel øker, minker den andre (negativ sammenheng).
- En verdi nær 0 betyr ingen klar sammenheng.

### Hvorfor heatmap?
Et korrelasjonsheatmap gjør det lett å:
- Identifisere hvilke variabler som henger tett sammen.
- Avdekke redundans eller potensielt interessante mønstre.
- Forstå hvordan værforhold henger sammen over tid.

Verdiene vises direkte i rutenettet for presis analyse.


```python
# Oppretter et objekt med sti til datasettet
plotter = VærDataPlotter("../data/R_Sammenslaaing_uten_feil.csv")

plotter.tegn_korrelasjonsheatmap()
```
